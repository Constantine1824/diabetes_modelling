```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import zipfile
import os

ZIPPED_FOLDER_PATH = "./Data/diabetes-health-indicators-dataset.zip"
ZIP_EXTRACT_PATH = "./Data"

def extract(zfile, extract_path):
    file = None
    with zipfile.ZipFile(zfile,  'r') as zf:
        file = zf.namelist()[0]
        #Check if file already exists
        if os.path.exists(os.path.join(extract_path, file)):
            return os.path.join(extract_path, file)
        zf.extractall(extract_path)
    print(f'Successful extraction of {file} to {extract_path}')
    return os.path.join(extract_path, file)

DATA_PATH = extract(ZIPPED_FOLDER_PATH, ZIP_EXTRACT_PATH)
```

# Feature Engineering and Selection

```python
df = data.copy()
y = df['diagnosed_diabetes']
X = df.drop(['diagnosed_diabetes', 'insulin_level', 'diabetes_stage'], axis=1)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

s_split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for i , (train_idx, test_idx) in enumerate(s_split.split(X, y)):
    X_train, y_train = X.loc[train_idx], y.loc[train_idx]
    X_test, y_test = X.loc[test_idx], y.loc[test_idx]
```

```python
from sklearn.base import TransformerMixin, BaseEstimator
sleep_idx, screen_idx = 'sleep_hours_per_day', 'screen_time_hours_per_day'

class NumericalTransformer(BaseEstimator, TransformerMixin):

    def __init__(self, skew_threshold=0.75, skew_transform=True):
        self.col = []
        self.skew_transform = skew_transform
        self.skew_threshold = skew_threshold
        self.skewed_features_ = []
        self.binary_features = ['family_history_diabetes', 'hypertension_history', 'cardiovascular_history'] 
        
    def fit(self, X, y=None):
        if isinstance(X, np.ndarray):
            X = pd.DataFrame(X)
        skewness = X.skew()
        self.skewed_features_ = skewness[abs(skewness) > self.skew_threshold].index.tolist()
        return self
        
    def transform(self, X):
        #self.col = self.col.append(X.columns)
        if isinstance(X, np.ndarray):
            X = pd.DataFrame(X)
        X = X.copy()
        X['free_hours_per_week'] = 24 - (X[sleep_idx] + X[screen_idx])#)
        if self.skew_transform:
            X = self.determine_skewness(X)
        return X

    def determine_skewness(self, X):
        for feature in self.skewed_features_:
            if feature in X.columns and feature not in self.binary_features:
                X[feature] = np.log1p(X[feature])
            return X
    
    def get_feature_names_out(self, input_features=None):
        print(type(input_features))
        # if not input_features.any():
        #     return self.col + ['free_hours_per_week']
        return list(input_features) + ['free_hours_per_week']
```

```python
numeric_df = X_train.select_dtypes(include=[np.number])
str_df = X_train.select_dtypes(exclude=[np.number])

numeric_col = list(numeric_df)
str_col = list(str_df)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

numeric_pipeline = Pipeline([
    ('skewtransform', NumericalTransformer()),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer([
    ('numeric', numeric_pipeline, numeric_col),
    ('encoder', OneHotEncoder(), str_col)
])
```

```python
from sklearn.ensemble import RandomForestClassifier

prep = preprocessor.fit(X_train)
X_rf = preprocessor.fit_transform(X_train)
rf = RandomForestClassifier(random_state=42)
rf.fit(X_rf, y_train)
feat_imp = pd.Series(rf.feature_importances_, index=prep.get_feature_names_out())
print(feat_imp.sort_values(ascending=False))
```

```python
from sklearn.linear_model import SGDClassifier, LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.feature_selection import SelectFromModel

rfs = RandomForestClassifier(n_estimators = 50, random_state=42)
sgd = Pipeline([
    ('transformers',preprocessor),
    ('feature_selector', SelectFromModel(rfs, threshold=0.14)),
    ('model', SGDClassifier())
])
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)
pretty_print_results(y_pred, y_test)
```

```python
log_clf = Pipeline([
    ('transformers', preprocessor),
    ('feature_selector', SelectFromModel(rfs, threshold=0.14)),
    ('model', LogisticRegression())
])
log_clf.fit(X_train, y_train)
y_pred = log_clf.predict(X_test)
pretty_print_results(y_pred, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = Pipeline([
    ('transformers', preprocessor),
    ('feature_selector', SelectFromModel(rfs, threshold='median')),
    ('model', RandomForestClassifier(n_jobs=-1, max_depth=12, n_estimators=200)),
])
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
pretty_print_results(y_pred, y_test)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_curve, auc

xgb = Pipeline([
    ('transformers', preprocessor),
    ('feature_selector', SelectFromModel(rfs, threshold='median')),
    ('model', XGBClassifier(n_estimators=50, random_state=42))
])
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
y_pred_proba = xgb.predict_proba(X_test)
fpr, tpr, threshold = roc_curve(y_test, y_pred_proba[:, 1])
pretty_print_results(y_pred, y_test)
auc(fpr, tpr)
```

# Model Performance Evaluation

```python
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import cross_val_predict

X_df = preprocessor.fit_transform(X_train)
rf2 = RandomForestClassifier(random_state=42)
y_cross_pred = cross_val_predict(rf2, X_df, y_train, cv=3, method='predict_proba')
```

```python
xgb2 = XGBClassifier(random_state=42, n_estimators=30)
xgb_pred = cross_val_predict(xgb2, X_df, y_train, cv=3, method='predict_proba')
```

```python
plt.figure(figsize=(7,5))
for model in [y_cross_pred[:, 1], xgb_pred[:, 1]]:
    fpr, tpr, threshold = roc_curve(y_train, model)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'AUC: {roc_auc:.2f}')

plt.plot([0, 1], [0, 1], 'r--', label='Random Guess')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves for Two Models')
plt.legend()
plt.show()
```

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.grid((10, 6))
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
y_cross_pred[:, -1]
```

```python
fpr, tpr, thresholds = roc_curve(y_train, y_cross_pred[:, 1])
plot_roc_curve(fpr, tpr, label='RandomForest')
```

# Model Finetuning
`This is to explore further if accuracy and precision can be improved by tuning the hyperparameters`

```python
from sklearn.model_selection import GridSearchCV
param_grid = [
    {
        'transformers__numeric__skewtransform__skew_transform' : [True, False], 
        'transformers__numeric__skewtransform__skew_threshold' : [0.5, 0.75, 0.8, 0.9, 1.0, 1.5], 
        'model__n_estimators' : [100, 200] 
    } 
] 
rf2 = Pipeline([ 
    ('transformers', preprocessor), 
    ('feature_selector', SelectFromModel(rfs, threshold='median')), 
    ('model', RandomForestClassifier(n_jobs=-1, max_depth=12, n_estimators=100)), 
]) 
grid = GridSearchCV(rf2, param_grid, cv=3, return_train_score=True, scoring='accuracy') 
grid.fit(X_train, y_train)
```

```python
cves = grid.cv_results_
for mean_score, params in zip(cves['mean_test_score'], cves['params']):
    print(mean_score, params)
```

```python
grid.best_params_
```

```python
y_pred = grid.best_estimator_.predict(X_test)
pretty_print_results(y_test, y_pred)
```

```python
y_pred = grid.best_estimator_.predict(X_train)
pretty_print_results(y_train, y_pred)
```

```python
y_proba = grid.best_estimator_.predict_proba(X_train)
y_proba = y_proba[:,-1]
```

```python
fpr, tpr, threshold = roc_curve(y_train, y_proba)
roc_auc = auc(fpr, tpr)
print(roc_auc)
plt.plot(fpr, tpr, label=f'AUC: {roc_auc:.2f}')
```

```python
y_true = (y_train == 1)
precision, recall, thresholds = precision_recall_curve(y_true, y_proba)

plt.plot(thresholds, precision[:-1],'g--', label='Precision')
plt.plot(thresholds, recall[:-1], 'b-', label='Recall')
plt.legend()
plt.grid((12,8))
plt.tight_layout()
plt.title("Precision VS Recall")
plt.show()
```

```python
threshold = thresholds[np.argmax(precision >=0.92)]
print(f'Threshold: {threshold}')
y_threshold_pred = (y_proba >= threshold)
pretty_print_results(y_true, y_threshold_pred)
```
